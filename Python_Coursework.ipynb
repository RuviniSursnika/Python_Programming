```python
#Task 01

import numpy as np  #It provides support in mathematical, scientific, engineering, and data science programming
import pandas as pd #It is for handling this real-world messy data

#To access data from the CSV file, we require a function read_csv() that retrieves data in the form of the Dataframe
DataFrame = pd.read_csv('C:/Users/Ruvini Suranika/Desktop/Greenwich/Term 01/Python/Coursework/Iris.csv')

#Sometimes csv file has null values, which are later displayed as NaN in Data Frame.
#Pandas dropna() method allows the user to analyze and drop Rows/Columns with Null values in different ways
MyDataFrame = DataFrame.dropna()

MyDataFrame.dropna(inplace = True) #inplace: It is a boolean which makes the changes in data frame itself if True
print("\nIris Dataset:\n".title())
print(MyDataFrame.to_string()) #To render the given DataFrame to a console-friendly tabular output

#To display the first 10 rows from the dataset
result = MyDataFrame.head(10)
print("\nFirst 10 rows of the DataFrame:\n")
print(result)

# By considering all species values - Measures of Central Tendency (Mean, Median & Mode) for Petal width

#Mean
Mean=round(MyDataFrame["Petal width"].mean(),2)
print(f'\nBy considering all species, Mean value is -> \n{Mean} \n')

#Median
Median = round(MyDataFrame["Petal width"].median(),2)
print(f'By considering all species, Median value is -> \n{Median} \n') 

#Mode
Mode = MyDataFrame["Petal width"].mode()
print(f'By considering all species, Mode value is -> \n{Mode} \n') 

# By considering all species values - Measures of Dispersion (Standard deviation & Range) for Petal width

#Standard Deviation
Standard_Deviation = round(MyDataFrame["Petal width"].std(),2)
print(f'By considering all species, Standard Deviation is -> \n{Standard_Deviation} \n')

#Maximum value 
Max=MyDataFrame["Petal width"].max()
print(f'Maximum value is -> \n{Max}')

#Minimum value
Min=MyDataFrame["Petal width"].min()
print(f'Minimum value is -> \n{Min}')

#Range
Range = max(MyDataFrame["Petal width"])-min(MyDataFrame["Petal width"])
print(f'By considering all the species, Range is ->\n{Range}\n')


#Extarcting Species counts
species_value_counts = MyDataFrame["Species"].value_counts() #to find the unique value counts of each element in the given object
print('Species value counts are:')
print(species_value_counts)

#By considering Setosa Species values - Extracting Setosa species values - Get rows with particular value
Setosa_data = MyDataFrame.loc[MyDataFrame["Species"] == "Setosa"]                       
print('\nSpecies of Setosa values for Petal width are: \n')
print(Setosa_data)

#By considering Petal width for Setosa Species values- Measures of Central Tendency (Mean, Median & Mode)

#Mean
Setosa_Mean = round(Setosa_data["Petal width"].mean(),2)
print(f'\nMean value of Setosa Species for Petal width is -> \n{Setosa_Mean} \n')

#Median
Setosa_Median = round(Setosa_data["Petal width"].median(),2)
print(f'Median value of Setosa Species for Petal width is -> \n{Setosa_Median} \n') 

#Mode
Setosa_Mode = Setosa_data["Petal width"].mode()
print(f'Mode value of Setosa Species for Petal width is -> \n{Setosa_Mode} \n')


#By considering Petal width for Setosa Species values - Measures of Dispersion (Standard deviation & Range)

#Standard Deviation
Setosa_std = round(Setosa_data["Petal width"].std(),2)
print(f'Standard Deviation of Setosa Species for Petal width is -> \n{Setosa_std} \n')

#Maximum value
Setosa_Max = Setosa_data["Petal width"].max()
print(f'\nMaximum value of Setosa Species for Petal width is ->\n{Setosa_Max}')

#Minimum value
Setosa_Min = Setosa_data["Petal width"].min()
print(f'Minimum value of Setosa Species for Petal width is -> \n{Setosa_Min}')

#Range
Setosa_Range = max(Setosa_data["Petal width"])-min(Setosa_data["Petal width"])
print(f'Range of Setosa Species for Petal width is ->\n{Setosa_Range}')

#By considering Versicolor Species values - Extracting Versicolor species values - Get rows with particular value
Versicolor_data = MyDataFrame.loc[MyDataFrame["Species"] == "Versicolor"]                       
print('\nSpecies of Versicolor values for Petal width are: \n')
print(Versicolor_data)


#By considering Petal width for Versicolor Species values- Measures of Central Tendency (Mean, Median & Mode)

#Mean
Versicolor_Mean = round(Versicolor_data["Petal width"].mean(),2)
print(f'\nMean value of Versicolor Species for Petal width is -> \n{Versicolor_Mean} \n')

#Median
Versicolor_Median = round(Versicolor_data["Petal width"].median(),2)
print(f'Median value of Versicolor Species for Petal width is -> \n{Versicolor_Median} \n') 

#Mode
Versicolor_Mode = Versicolor_data["Petal width"].mode()
print(f'Mode value of Versicolor Species for Petal width is -> \n{Versicolor_Mode} \n')

#By considering Petal width for Versicolor Species values - Measures of Dispersion (Standard deviation & Range)

#Standard Deviation
Versicolor_std = round(Versicolor_data["Petal width"].std(),2)
print(f'Standard Deviation of Versicolor Species for Petal width is -> \n{Versicolor_std} \n')

#Maximum value
Versicolor_Max = Versicolor_data["Petal width"].max()
print(f'\nMaximum value of Versicolor Species for Petal width is ->\n{Versicolor_Max}')

#Minimum value
Versicolor_Min = Versicolor_data["Petal width"].min()
print(f'Minimum value of Versicolor Species for Petal width is -> \n{Versicolor_Min}')

#Range
Versicolor_Range = max(Versicolor_data["Petal width"])-min(Versicolor_data["Petal width"])
print(f'Range of Versicolor Species for Petal width is ->\n{Versicolor_Range}')

#By considering Petal width for Virginica Species values - Extracting Virginica species values - Get rows with particular value
Virginica_data = MyDataFrame.loc[MyDataFrame["Species"] == "Virginica"]                       
print('\nSpecies of Virginica values for Petal width are: \n')
print(Virginica_data)

#By considering Petal width for Virginica Species values- Measures of Central Tendency (Mean, Median & Mode)

#Mean
Virginica_Mean = round(Virginica_data["Petal width"].mean(),2)
print(f'\nMean value of Virginica Species for Petal width is -> \n{Virginica_Mean} \n')

#Median
Virginica_Median = round(Virginica_data["Petal width"].median(),2)
print(f'Median value of Virginica Species for Petal width is -> \n{Virginica_Median} \n') 

#Mode
Virginica_Mode = Virginica_data["Petal width"].mode()
print(f'Mode value of Virginica Species for Petal width is -> \n{Virginica_Mode} \n')


#By considering Petal width for Virginica Species values - Measures of Dispersion (Standard deviation & Range)

#Standard Deviation
Virginica_std = round(Virginica_data["Petal width"].std(),2)
print(f'Standard Deviation of Virginica Species for Petal width is -> \n{Virginica_std} \n')

#Maximum value
Virginica_Max = Virginica_data["Petal width"].max()
print(f'\nMaximum value of Virginica Species for Petal width is ->\n{Virginica_Max}')

#Minimum value
Virginica_Min = Virginica_data["Petal width"].min()
print(f'Minimum value of Virginica Species for Petal width is -> \n{Virginica_Min}')

#Range
Virginica_Range = max(Virginica_data["Petal width"])-min(Virginica_data["Petal width"])
print(f'Range of Virginica Species for Petal width is ->\n{Virginica_Range}')
```

```python
#Task 03

#with existing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Declare variables for the number of samples and the array which will store the dataset
NumberOfSamples = int(150)
MyArray = np.empty(shape=(NumberOfSamples, 4))

# Dimension 1: Values for Sepal length as array
Sepal_Length = MyDataFrame['Sepal length'].to_numpy()
for i in range(0, NumberOfSamples):
    MyArray[i, 0] = Sepal_Length[i]
    
#Dimension 2: Values for Sepal width as array   
Sepal_Width = MyDataFrame['Sepal width'].to_numpy()
for i in range(0, NumberOfSamples):
    MyArray[i, 1] = Sepal_Width[i]

#Dimension 3: Values for Petal length as array
Petal_Length = MyDataFrame['Petal length'].to_numpy()
for i in range(0, NumberOfSamples):
    MyArray[i, 2] = Petal_Length[i]

#Dimension 4: Values for Petal width as array
Petal_Width = MyDataFrame['Petal width'].to_numpy()
for i in range(0, NumberOfSamples):
    MyArray[i, 3] = Petal_Width[i]

#Species attribute names
Name = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

#Visualise the spines (axis) of the parallel coordinates
for i in range(0, 5):
    plt.vlines(i, 0, 10, '#808080')

#Generate the parallel coordinates diagram
for i in range(0, NumberOfSamples):
    plt.plot(Name, MyArray[i])
    
plt.title("Parallel Coordinate Diagram with existing libraries\n ") #Title of the diagram

#Render the plot
plt.show()

```

```python
# with new libraries

import pandas
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates #newly added library for the paralle coordinate diagram

# import the iris dataset
import seaborn as sns
data = sns.load_dataset('iris')

# Make the plot
parallel_coordinates(data, 'species', colormap=plt.get_cmap("Set2")) #Parallel coordinate diagram
plt.title("Parallel Coordinate Diagram with new library\n ") #Title of the diagram
plt.xlabel('\nSpecies\n') #Label of the X-axis of the graph
plt.ylabel('\nAttribute values\n') #Label of the Y-axis of the graph
plt.show()
```
